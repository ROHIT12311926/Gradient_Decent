```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_targets=1,noise=20)
```

```python
sns.scatterplot(x=X.ravel(),y=y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
sns.scatterplot(x=X.ravel(),y=y)
sns.lineplot(x=X.ravel(),y=lr.predict(X),c='r')
```

```python
class GDregressor:
    def __init__(self,learning_rate,epochs):
        self.m=64.39891532
        self.b=-120
        self.learning_rate=learning_rate
        self.epochs=epochs


    def fit(self,X,y):
        sns.scatterplot(x=X.ravel(),y=y)
        

        for i in range(self.epochs):
            slope=-2*np.sum(y-self.m*X.ravel()-self.b)
            self.b=self.b-(self.learning_rate*slope)
            y_pred=self.m*X.ravel()+self.b

            print(self.b)

            sns.lineplot(x=X.ravel(),y=y_pred)
            
            
            
        
        print(self.b)

    

```

```python
gdr=GDregressor(0.001,100)
```
