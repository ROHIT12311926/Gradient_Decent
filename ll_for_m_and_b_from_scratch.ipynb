```python

losses=[]
all_m=[]
all_b=[]
class GradientRegressor:
    def __init__(self,m,b,epochs,learning_rate):
        self.m=m
        self.b=b
        self.epochs=epochs
        self.learning_rate=learning_rate


    def fit(self,X,y):
        
        for i in range (self.epochs):
            slope_b=-2*np.sum(y-self.m*X.ravel()-self.b)
            slope_m=-2*np.sum((y-self.m*X.ravel()-self.b)*X.ravel())

            self.b=self.b-self.learning_rate*slope_b
            self.m=self.m-self.learning_rate*slope_m
            y_pred=self.m*X.ravel()+self.b

            loss=np.sum((y-self.m*X.ravel()-self.b)*(y-self.m*X.ravel()-self.b))

            losses.append(loss)
            all_m.append(self.m)
            all_b.append(self.b)
            sns.scatterplot(x=X.ravel(),y=y)
            sns.lineplot(x=X.ravel(),y=y_pred)

            
        print(self.b,self.m)

    

        
```

```python
gdr=GradientRegressor(1,1,100,0.001)
```

```python
m_arr = np.linspace(-150, 150, 10)
b_arr = np.linspace(-150, 150, 10)
mGrid, bGrid = np.meshgrid(m_arr,b_arr)

final = np.vstack((mGrid.ravel().reshape(1,100),bGrid.ravel().reshape(1,100))).T

z_arr = []

for i in range(final.shape[0]):
    z_arr.append(np.sum((y - final[i,0]*X.reshape(100) - final[i,1])**2))

z_arr = np.array(z_arr).reshape(10,10)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(x = m_arr, y = b_arr, z =z_arr)])

fig.update_layout(title='Cost Function', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))



fig.show()

fig.write_html("cost_function.html")
```
