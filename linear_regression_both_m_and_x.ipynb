```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_features=1,n_targets=1,noise=20)
```

```python
X
```

```python
sns.scatterplot(x=X.ravel(),y=y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
sns.scatterplot(x=X.ravel(),y=y)
sns.lineplot(x=X.ravel(),y=lr.predict(X),c='r')
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
m=1
b=1
sns.scatterplot(x=X.ravel(),y=y)
sns.lineplot(x=X.ravel(),y=lr.predict(X),c='r')


for i in range (1000):
    slope_b=-2*np.sum(y-m*X.ravel()-b)
    slope_m=-2*np.sum((y-m*X.ravel()-b)*X.ravel())

    b=b-0.001*slope_b

    m=m-0.001*slope_m

    y_pred=m*X.ravel()+b


    sns.lineplot(x=X.ravel(),y=y_pred)
    

print(b)
print(m)




    




    
```
