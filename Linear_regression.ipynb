```python
import pandas as pd
import pandas as pd
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_targets=1,random_state=13,n_informative=1,noise=80)
```

```python
X
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=X.ravel(),y=y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.intercept_
```

```python
sns.scatterplot(x=X.ravel(),y=y)
sns.lineplot(x=X.ravel(),y=lr.predict(X),c='r')
```

```python
m=lr.coef_
```

```python
y_pred=m*X+0
```

```python
y_pred
```

```python
sns.scatterplot(x=X.ravel(),y=y)
sns.lineplot(x=X.ravel(),y=lr.predict(X),c='r')
sns.lineplot(x=X.ravel(),y=y_pred.ravel(),c='green')
```

```python
b=10
epochs=200
slope=-2 * np.sum(y-m*X.ravel()-b)
sns.lineplot(x=X.ravel(),y=lr.predict(X),c='r')


for i in range (epochs):
    colors=sns.color_palette("husl",epochs)
    slope=-2 * np.sum(y-m*X.ravel()-b)
    y_pred=m*X.ravel()+b
    b=b-0.01*slope

    sns.scatterplot(x=X.ravel(),y=y)
    
    sns.lineplot(x=X.ravel(),y=y_pred.ravel(),color=colors[i])
    print(b+(0.01*slope))
```
